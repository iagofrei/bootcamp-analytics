10.  Utilizando uma query em SQL e o banco `sqlite-sakila.db` retorne as cidades e os países, mas apenas os países que começem com a letra **B**.

```python
from sqlalchemy import create_engine
from sqlalchemy import inspect
```

```python
engine = create_engine('sqlite:///atividade1/sqlite-sakila.db')
insp = inspect(engine)
```

```python
insp.get_table_names()
```

```python
sql_query = '''
select
    *
from
    country
'''
```

```python
paises = pd.read_sql_query(sql_query, engine)
paises
```

```python
sql_query_2 = '''
select
    *
from
    city
'''

cidades = pd.read_sql_query(sql_query_2, engine)
cidades
```

```python
sql_query_3 ='''
SELECT
    country.country,
    city.city
FROM country
INNER JOIN city
ON country.country_id = city.country_id
WHERE country.country LIKE 'B%'
 '''
paises_cidades_df = pd.read_sql_query(sql_query_3, engine)
paises_cidades_df
```
